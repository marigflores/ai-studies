```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Dados de exemplo
data = {
    'Característica 1': [1.2, 2.3, 3.4],
    'Característica 2': [3.4, 4.5, 5.6],
    'Característica 3': [5.6, 6.7, 7.8],
    'Característica 4': [7.8, 8.9, 9.0],
    'Característica 5': [9.0, 0.1, 1.2],
}
df = pd.DataFrame(data)
```

```python
# Padronizando os dados
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
# Aplicando PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(df_scaled)
```

```python
# Criando um DataFrame com os componentes principais
df_pca = pd.DataFrame(data=principalComponents, columns=['Principal Component 1', 'Principal Component 2'])
```

```python
# Visualizando os componentes principais
plt.scatter(df_pca['Principal Component 1'], df_pca['Principal Component 2'])
plt.title('PCA dos Dados')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.show()
```

```python
# Exibindo a variância explicada por cada componente
print('Variância explicada por cada componente:', pca.explained_variance_ratio_)
```
